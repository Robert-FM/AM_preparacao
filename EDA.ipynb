```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
#Gerando alguns dados de exemplo (Gaussiano)
dados_histograma = np.random.normal(loc=0,scale=1,size=1000)

#Criando o histograma
plt.figure(figsize=(8,6))
plt.hist(dados_histograma,bins=30,edgecolor='black',alpha=0.7,label='cdz')
plt.title('Histograma dos Dados')
plt.xlabel('Valores')
plt.ylabel('Frequência')
plt.legend()
plt.show()
```

```python
#Gerando alguns dados de exemplo para o scatter plot
x = np.random.rand(50)
y = 2*x+np.random.rand(50) # Criando uma relação linear com algum ruído

#Criando o scatter plot
plt.figure(figsize=(8,6))
plt.scatter(x,y,color='blue', marker='o',alpha=0.8,label='cdz')
plt.title('Scatter Plot da Relação entre X e Y')
plt.xlabel('Variável X')
plt.ylabel('Variável Y')
plt.legend()
plt.show()
```

```python
# Gerando alguns dados de exemplo para o box plot (múltiplas distribuições)
dados_boxplot = [np.random.normal(loc=0, scale=1, size=100),
                 np.random.normal(loc=2, scale=1.5, size=120),
                 np.random.normal(loc=-1, scale=0.8, size=90)]

# Criando o box plot
plt.figure(figsize=(8, 6))
plt.boxplot(dados_boxplot, labels=['Grupo A', 'Grupo B', 'Grupo C'], patch_artist=True, boxprops=dict(facecolor='lightblue'),label='cdz')
plt.title('Box Plot de Múltiplos Grupos')
plt.ylabel('Valores')
plt.grid(axis='y', alpha=0.7)
plt.legend()
plt.show()
```

```python
# Definindo os parâmetros da função linear
m = 2  # Inclinação
b = 1  # Intercepto y

# Criando alguns valores de x para plotar a linha
x = np.linspace(0, 10, 100)  # Cria 100 pontos entre 0 e 10

# Calculando os valores de y correspondentes usando a função linear
y = m * x + b

# Criando o plot
plt.figure(figsize=(8, 6))
ax = plt.gca()  # Obtém os eixos correntes (current axes)

# Oculta as spines superior e direita
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.plot(x, y, color='red', linestyle='-', linewidth=2, label=f'y = {m}x + {b}')
plt.title('Gráfico de Função Linear (Sem Spines Superior e Direita)')
plt.xlabel('Variável X')
plt.ylabel('Variável Y')
plt.legend()
plt.xlim(0, 10)
plt.ylim(0, 25)
plt.xticks(np.arange(0, 11, 1))
plt.yticks(np.arange(0, 26, 5))
plt.show()
```

```python
# 1. Gerando dados para o gráfico de dispersão
np.random.seed(0)  # Para reprodutibilidade
x_scatter = np.random.rand(50) * 10  # Valores de x entre 0 e 10
y_scatter = 2 * x_scatter + 1 + np.random.randn(50) * 5  # Relação linear com ruído

# 2. Definindo os parâmetros da função linear
m = 2  # Inclinação (a mesma da relação subjacente aos dados de dispersão)
b = 1  # Intercepto y (o mesmo da relação subjacente aos dados de dispersão)

# Gerando pontos para a função linear (no mesmo intervalo de x dos dados de dispersão)
x_linear = np.linspace(0, 10, 100)
y_linear = m * x_linear + b

# 3. Criando a figura e os eixos
plt.figure(figsize=(10, 6))
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# 4. Plotando o gráfico de dispersão
plt.scatter(x_scatter, y_scatter, color='blue', alpha=0.7, label='Dados Observados')

# 5. Plotando a função linear
plt.plot(x_linear, y_linear, color='red', linestyle='-', linewidth=2, label=f'Função Linear: y = {m}x + {b}')

# 6. Adicionando títulos, rótulos e legenda
plt.title('Gráfico de Dispersão com Função Linear')
plt.xlabel('Variável X')
plt.ylabel('Variável Y')
plt.legend()

# 7. Exibindo o gráfico
plt.show()
```

```python
# Calcula a matriz de correlação
correlation_matrix = df.corr()

# Cria uma máscara para a parte superior
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))

# Configura o tamanho da figura
plt.figure(figsize=(16, 14))

# Cria o heatmap com a máscara
sns.heatmap(correlation_matrix, mask=mask, cmap='coolwarm', annot=True, fmt=".2f", linewidths=.5)

plt.title("Matriz de Correlação (sem duplicatas)")
plt.show()
```
