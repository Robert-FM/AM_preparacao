```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
#Gerando alguns dados de exemplo (Gaussiano)
dados_histograma = np.random.normal(loc=0,scale=1,size=1000)

#Criando o histograma
plt.figure(figsize=(8,6))
plt.hist(dados_histograma,bins=30,edgecolor='black',alpha=0.7,label='cdz')
plt.title('Histograma dos Dados')
plt.xlabel('Valores')
plt.ylabel('FrequÃªncia')
plt.legend()
plt.show()
```

```python
#Gerando alguns dados de exemplo para o scatter plot
x = np.random.rand(50)
y = 2*x+np.random.rand(50) # Criando uma relaÃ§Ã£o linear com algum ruÃ­do

#Criando o scatter plot
plt.figure(figsize=(8,6))
plt.scatter(x,y,color='blue', marker='o',alpha=0.8,label='cdz')
plt.title('Scatter Plot da RelaÃ§Ã£o entre X e Y')
plt.xlabel('VariÃ¡vel X')
plt.ylabel('VariÃ¡vel Y')
plt.legend()
plt.show()
```

```python
# Gerando alguns dados de exemplo para o box plot (mÃºltiplas distribuiÃ§Ãµes)
dados_boxplot = [np.random.normal(loc=0, scale=1, size=100),
                 np.random.normal(loc=2, scale=1.5, size=120),
                 np.random.normal(loc=-1, scale=0.8, size=90)]

# Criando o box plot
plt.figure(figsize=(8, 6))
plt.boxplot(dados_boxplot, labels=['Grupo A', 'Grupo B', 'Grupo C'], patch_artist=True, boxprops=dict(facecolor='lightblue'),label='cdz')
plt.title('Box Plot de MÃºltiplos Grupos')
plt.ylabel('Valores')
plt.grid(axis='y', alpha=0.7)
plt.legend()
plt.show()
```

```python
# Definindo os parÃ¢metros da funÃ§Ã£o linear
m = 2  # InclinaÃ§Ã£o
b = 1  # Intercepto y

# Criando alguns valores de x para plotar a linha
x = np.linspace(0, 10, 100)  # Cria 100 pontos entre 0 e 10

# Calculando os valores de y correspondentes usando a funÃ§Ã£o linear
y = m * x + b

# Criando o plot
plt.figure(figsize=(8, 6))
ax = plt.gca()  # ObtÃ©m os eixos correntes (current axes)

# Oculta as spines superior e direita
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.plot(x, y, color='red', linestyle='-', linewidth=2, label=f'y = {m}x + {b}')
plt.title('GrÃ¡fico de FunÃ§Ã£o Linear (Sem Spines Superior e Direita)')
plt.xlabel('VariÃ¡vel X')
plt.ylabel('VariÃ¡vel Y')
plt.legend()
plt.xlim(0, 10)
plt.ylim(0, 25)
plt.xticks(np.arange(0, 11, 1))
plt.yticks(np.arange(0, 26, 5))
plt.show()
```

```python
# 1. Gerando dados para o grÃ¡fico de dispersÃ£o
np.random.seed(0)  # Para reprodutibilidade
x_scatter = np.random.rand(50) * 10  # Valores de x entre 0 e 10
y_scatter = 2 * x_scatter + 1 + np.random.randn(50) * 5  # RelaÃ§Ã£o linear com ruÃ­do

# 2. Definindo os parÃ¢metros da funÃ§Ã£o linear
m = 2  # InclinaÃ§Ã£o (a mesma da relaÃ§Ã£o subjacente aos dados de dispersÃ£o)
b = 1  # Intercepto y (o mesmo da relaÃ§Ã£o subjacente aos dados de dispersÃ£o)

# Gerando pontos para a funÃ§Ã£o linear (no mesmo intervalo de x dos dados de dispersÃ£o)
x_linear = np.linspace(0, 10, 100)
y_linear = m * x_linear + b

# 3. Criando a figura e os eixos
plt.figure(figsize=(10, 6))
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# 4. Plotando o grÃ¡fico de dispersÃ£o
plt.scatter(x_scatter, y_scatter, color='blue', alpha=0.7, label='Dados Observados')

# 5. Plotando a funÃ§Ã£o linear
plt.plot(x_linear, y_linear, color='red', linestyle='-', linewidth=2, label=f'FunÃ§Ã£o Linear: y = {m}x + {b}')

# 6. Adicionando tÃ­tulos, rÃ³tulos e legenda
plt.title('GrÃ¡fico de DispersÃ£o com FunÃ§Ã£o Linear')
plt.xlabel('VariÃ¡vel X')
plt.ylabel('VariÃ¡vel Y')
plt.legend()

# 7. Exibindo o grÃ¡fico
plt.show()
```

```python
# Calcula a matriz de correlaÃ§Ã£o
correlation_matrix = df.corr()

# Cria uma mÃ¡scara para a parte superior
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))

# Configura o tamanho da figura
plt.figure(figsize=(16, 14))

# Cria o heatmap com a mÃ¡scara
sns.heatmap(correlation_matrix, mask=mask, cmap='coolwarm', annot=True, fmt=".2f", linewidths=.5)

plt.title("Matriz de CorrelaÃ§Ã£o (sem duplicatas)")
plt.show()
```

```python
# Carregar o arquivo Excel
arquivo_excel = 'Dados Capibaribe correlaÃ§Ã£o.xlsx'  # Substitua pelo caminho do seu arquivo
dados = pd.read_excel(arquivo_excel)
colunas_numericas = dados.select_dtypes(include=[np.number]).columns

# Iterar pelas colunas numÃ©ricas
for coluna in colunas_numericas:
    dados_coluna = dados[coluna].dropna()  # Remove valores ausentes (NaN)
    
    media = np.mean(dados_coluna)
    mediana = np.median(dados_coluna)
    desvio_padrao = np.std(dados_coluna)
    coeficiente_variacao = (desvio_padrao / media) * 100
    minimo = np.min(dados_coluna)
    maximo = np.max(dados_coluna)
    
    print(f"\nColuna: {coluna}")
    print(f"  MÃ©dia: {media:.2f}")
    print(f"  Mediana: {mediana:.2f}")
    print(f"  Desvio PadrÃ£o: {desvio_padrao:.2f}")
    print(f"  Coeficiente de VariaÃ§Ã£o: {coeficiente_variacao:.2f}")
    print(f"  MÃ­nimo: {minimo:.2f}")
    print(f"  MÃ¡ximo: {maximo:.2f}")
```
